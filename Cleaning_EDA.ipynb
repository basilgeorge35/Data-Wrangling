```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Balu\Desktop\Projects\Data_Wrangling\Customer Call List.xlsx")
```

```python
df
```

```python
df = df.drop_duplicates()
df
```

```python
df.drop(columns = 'Not_Useful_Column', inplace = True)
df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("/..._")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^0-9]','')
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x : str(x))
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('--','')
```

```python
df
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', 2, expand = True)
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df['Paying Customer'] = df['Paying Customer'].str.replace('N/a','N')
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df
```
